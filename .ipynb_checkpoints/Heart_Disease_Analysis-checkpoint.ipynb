## Step 3.1: Basic Dataset information


---

| Variable | Name / Description | Values / Interpretation |
|----------|------------------|------------------------|
| **age** | Age of the patient | Continuous (years) |
| **sex** | Sex of the patient | 0 = female, 1 = male |
| **chest_pain** | Chest pain type | 0 = typical angina<br>1 = atypical angina<br>2 = non-anginal pain<br>3 = asymptomatic |
| **resting_bp** | Resting blood pressure | Continuous (mm Hg) |
| **chol** | Serum cholesterol | Continuous (mg/dl) |
| **fbs** | Fasting blood sugar | 0 = ≤120 mg/dl, 1 = >120 mg/dl |
| **resting_ecg** | Resting electrocardiographic results | 0 = normal<br>1 = ST-T wave abnormality<br>2 = probable/definite LV hypertrophy |
| **thalach** | Maximum heart rate achieved | Continuous (bpm) |
| **ex_ang** | Exercise-induced angina | 0 = no, 1 = yes |
| **old_peak** | ST depression induced by exercise | Continuous (numeric) |
| **slope** | Slope of peak exercise ST segment | 0 = upsloping<br>1 = flat<br>2 = downsloping |
| **ca** | Major vessels colored by fluoroscopy | Integer (0–3) |
| **thal** | Thalassemia | 0 = normal<br>1 = fixed defect<br>2 = reversible defect |
| **heart_disease** | Presence of heart disease | 0 = no, 1 = yes |

To understand how stroke risk varies across demographic and health factors, I analyzed the distribution of categorical features in our dataset plotting frequency distribution of categories and proportion of stroke cases  within each category

```python
fig, axes = plt.subplots(len(cate_vars), 2, figsize=(10, 35), dpi=150)

# Define a color palette (you can choose any palette you like)
palette = sns.color_palette("husl", n_colors=len(cate_vars)*5)  # Generous number of colors

for i, var in enumerate(cate_vars):
    label = var.replace('_' , ' ').title()
    # Left plot: seaborn countplot with different colors for each group
    sns.countplot(
        x=var,
        data=heart_data,
        ax=axes[i, 0],
        palette=palette,
        hue = var,
        edgecolor='black',
        legend = False
    )
    axes[i, 0].grid(True, linestyle='-', color='gray', linewidth=0.7, alpha=0.5, zorder=0)
    axes[i, 0].set_facecolor('#f0f5ff')
    axes[i , 0].set_xlabel(label)
    axes[i , 0].set_ylabel('Count')
    axes[i, 0].tick_params(axis='x', rotation=45)

    # Raise bars above grid
    for patch in axes[i, 0].patches:
        patch.set_zorder(3)

    # Right plot: stacked bar normalized by row (unchanged)
    label = var.replace('_', ' ').title()
    data = heart_data.groupby([var, 'heart_disease']).size().unstack(fill_value=0)
    data_norm = data.div(data.sum(axis=1), axis=0)  # Normalize Data

    data_norm.plot(
        kind='bar',
        stacked=True,
        ax=axes[i, 1],
        color=['lightblue', 'steelblue'],
        edgecolor='black'
    )
    axes[i, 1].set_ylabel(label)
    axes[i, 1].legend(
        title='Heart Disease',
        labels=['No', 'Yes'],
        loc='upper left',
        bbox_to_anchor=(1, 1)
    )
    axes[i , 1].set_xlabel(label)
    axes[i , 1].set_ylabel('Proportion')
    axes[i, 1].grid(True, linestyle='-', color='gray', linewidth=0.7, alpha=0.5, zorder=0)
    axes[i, 1].set_facecolor('#f0f5ff')
    axes[i, 1].tick_params(axis='x', rotation=45)

    # Raise bars above grid on right plot
    for patch in axes[i, 1].patches:
        patch.set_zorder(3)

plt.tight_layout()
plt.show()
```

# Step 5: Data Preprocessing



---

## Step 5.1: Feature Encoding

---
Due to machine learning models mainly only being able to take numerical input we will one-hot encode our categorical variables as they are either binary or with > 2 groups but the groups have no order for example the the work type has no inherent order the groups are just different. We will map gender , eve_married , and residence_type to 0 and 1 , and we will one-hot encode work_type and smoking_status


---

```python
heart_data = pd.read_csv("/content/drive/MyDrive/DATASETS/heart.csv")

heart_data = pd.get_dummies(
    heart_data,
    columns=['restecg' , 'slope', 'ca' , 'thal'],
    drop_first=True,
    dtype=int
)

heart_data = heart_data.rename(columns={'target': 'heart_disease'})
```

## Step 5.2: Define Test/Train Split

---
We will use a 80/20 test / train split on the data to ensure that we can test the model on unseen data and we will define now as before we deal with missing values or do transformations we will split to ensure there is no data leakage. We will drop 'id' as it is a unique identifier and provides nothing to analysis


---

```python
X = heart_data.drop(['heart_disease'] , axis = 1)
y = heart_data['heart_disease']

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25 , stratify = y , random_state = 30)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Step 6: Methods


To predict heart disease we are going to use logistic regression , a decision tree and a random forest. We will implement 5-fold cross validation to tune the hyperparameters. We will look at several scoring metrics but will put the most priority on recall as that tells us of all people with heart disease how many did we accurately predict as having heart disease as that is the most important goal of this analysis

# Step 7: Modeling

## Step 7.1: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV


param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 3],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2]
}


dt = DecisionTreeClassifier(random_state=324)


grid_dt = GridSearchCV(
    estimator=dt,
    param_grid=param_grid_dt,
    cv=5,
    scoring='recall',
    n_jobs=-1
)


grid_dt.fit(X_train, y_train)


best_dt = grid_dt.best_estimator_
print("Best params:", grid_dt.best_params_)
print("Best CV recall:", grid_dt.best_score_)


y_pred = best_dt.predict(X_train)
print("\nTest Set Classification Report:")
print(classification_report(y_train, y_pred))

y_pred = best_dt.predict(X_test)
print("\nTest Set Classification Report:")
print(classification_report(y_test, y_pred))




```

## Step 7.2: Logistic Regression

```python
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],   # regularization strength
    'penalty': ['l2'],              # l2 only for simplicity
    'solver': ['lbfgs'],            # lbfgs supports l2
    'max_iter': [100, 200, 500,1000]
}

# base model
lr = LogisticRegression(random_state=324)

# grid search with recall
grid_lr = GridSearchCV(
    estimator=lr,
    param_grid=param_grid_lr,
    cv=5,
    scoring='recall',
    n_jobs=-1
)


grid_lr.fit(X_train_scaled, y_train)


best_lr = grid_lr.best_estimator_
print("Best params:", grid_lr.best_params_)
print("Best CV recall:", grid_lr.best_score_)


y_pred = best_lr.predict(X_train_scaled)
print("\nTrain Set Classification Report:")
print(classification_report(y_train, y_pred))


y_pred = best_lr.predict(X_test_scaled)
print("\nTest Set Classification Report:")
print(classification_report(y_test, y_pred))
```

## Step 7.3: Random Forest

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score

# Create pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(random_state=324))
])

# Correct param grid for pipeline
param_grid_svm = {
    'svm__C': [0.005, 0.01, 0.05, 0.1, 1, 10, 20],
    'svm__kernel': ['linear', 'rbf', 'poly'],
    'svm__gamma': ['scale', 'auto', 0.1, 0.5, 1, 5],
    'svm__degree': [2, 3, 4]
}

# Grid search over pipeline
grid_svc = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid_svm,
    cv=5,
    scoring='accuracy',  # balances recall for both classes
    n_jobs=-1
)

# Fit the grid search
grid_svc.fit(X_train, y_train)

# Use the best estimator
best_svc = grid_svc.best_estimator_
print("Best params:", grid_svc.best_params_)
print("Best CV recall:", grid_svc.best_score_)

# Train predictions
y_pred_train = best_svc.predict(X_train)
print("\nTrain Set Classification Report:")
print(classification_report(y_train, y_pred_train))

# Test predictions
y_pred_test = best_svc.predict(X_test)
print("\nTest Set Classification Report:")
print(classification_report(y_test, y_pred_test))



```

| Model           | Train Accuracy | Test Accuracy | Train Recall | Test Recall | Train F1-score | Test F1-score |

| Random Forest   |0.89                | 0.86              | 0.94             |0.93             |0.91                | 0.87              |
